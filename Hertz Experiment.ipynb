```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly
import numpy as np
```

```python
# Data
Voltage = []
Ampere = []
while True:  
    V = input("Voltage:\n")
    if V == '#':
        break
    A = input("Ampere:\n")
    Voltage.append(float(V))
    Ampere.append(float(A))
```

```python
data_csv = pd.DataFrame({"V": Voltage, "A": Ampere})
data_csv.to_csv("data_Hertz")
```

```python
data = pd.read_csv('data_Hertz')
plt.plot(data.V, data.A)
plt.xlabel("V", size = 32)
plt.ylabel("A", size = 32)
```
